## 1.2 Complete graph

```python
def complete_graph(n):
    A = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            if i!=j:
                A[i,j] = 1
                A[j,i] = 1
    L=np.diag(A.sum(axis=0))-A
    return L
```

```python
n=10
L = complete_graph(n)
t_perturbed = [0.001,0.01,0.1,1,10,100,1000]
all_val,all_vec = maximal_perturbation(L,n-1,t_perturbed)
```

```python
def eigenpair(L):
    val,vec=np.linalg.eig(L) 
    indices = np.argsort(val)
    val = val[indices]
    vec = vec[:,indices]
    vmax = np.argmax(vec[:,1])
    vmin = np.argmin(vec[:,1])
    return val, vec, vmin, vmax
```

```python
def perturbed_eigenpair(G,v,weight):
    n= len(G.nodes)
    Gp = G.copy() 
    Gp.add_edge(v,n,weight=weight)
    Lp = nx.laplacian_matrix(Gp)
    Lp = Lp.toarray()
    '''valp,vecp=np.linalg.eig(Lp) 
    indices = np.argsort(valp)
    valp = valp[indices]
    vecp = vecp[:,indices]'''
    res = eigenpair(Lp)
    return res[0], res[1], Gp
    #return valp,vecp,Gp
```

## 2.2 Quantitative comparison, Figure 3 right

```python
import time
```

```python
def maximal_perturbation2(L,node,tmin,tmax): # Dichotomic search
    n = len(L)
    while np.abs(tmax-tmin)>1e-2:
        t = np.sqrt(tmin*tmax)
        eigval,eigvec=maximal_perturbation(L,node,[t])
        eigvec[:,1,0] = eigvec[:,1,0] * np.sign(eigvec[n,1,0])
        if eigvec[n,1,0]<eigvec[:,1,0].max():
            tmax = t
        else:
            tmin = t
    return t

def threshold_centrality2(L,tmin,tmax):
    n = len(L)
    max_perturbation=np.zeros(n)
    threshold_t = np.zeros((n,))-1
    for node in range(n):
        threshold_t[node]=maximal_perturbation2(L,node,tmin,tmax)
    return threshold_t
```

```python
n = 20
N_iter = 100
all_m = np.arange(30,160,2)
all_results = np.zeros((len(all_m),6))

t0 = time.time()
for i,m in enumerate(all_m):
    #m = np.int(p*n*(n-1)/2)
    all_correlations = []
    for k in range(N_iter):
        G = nx.gnm_random_graph(n, m)
        centrality = nx.betweenness_centrality(G)
        b_cent = [v for v in centrality.values()]
        centrality = nx.closeness_centrality(G)
        c_cent = [v for v in centrality.values()]
        centrality = nx.eigenvector_centrality(G)
        e_cent = [v for v in centrality.values()]
        L = nx.laplacian_matrix(G)
        L = L.toarray()
        t_centrality = threshold_centrality2(L,1e-10,1e10)
        centralities = [1/t_centrality,b_cent,c_cent,e_cent]
        correlation = []
        for ii in range(len(centralities)):
            for jj in range(ii+1,len(centralities)):
                correlation.append(np.corrcoef(centralities[ii],centralities[jj])[0,1])
        all_correlations.append(correlation)
            #print(k)
    all_results[i,:] = np.nanmean(all_correlations,axis=0)
    print(m)
print("Total time = "+str(time.time()-t0))
```

```python
plt.figure(figsize=(9,5))
plt.plot(all_m,all_results)
plt.legend(['fcd/bet','fcd/clo','fdp/eig','bet/clo','bet/eig','clo/eig'])
plt.ylim([0,1])
plt.xlabel("m",fontsize = 16)
plt.ylabel("Correlation",fontsize = 16)
```

### Number of steps in the dichotomic search
This number is bounded by $\log_2(tmax/tmin)$

```python
def maximal_perturbation3(L,node,tmin,tmax): 
    # Dichotomic search => counting the number of steps <= log_2(tmax/tmin)
    n = len(L)
    nstep = 0
    while np.abs(tmax-tmin)>1e-2:
        nstep+=1
        t = np.sqrt(tmin*tmax)
        eigval,eigvec=maximal_perturbation(L,node,[t])
        eigvec[:,1,0] = eigvec[:,1,0] * np.sign(eigvec[n,1,0])
        if eigvec[n,1,0]<eigvec[:,1,0].max():
            tmax = t
        else:
            tmin = t
    return t, nstep

def threshold_centrality_comp_time(L,tmin,tmax):
    n = len(L)
    max_perturbation=np.zeros(n)
    threshold_t = np.zeros((n,))-1
    computation_time = np.zeros((n,))
    steps = np.zeros((n,))
    for node in range(n):
        t0= time.time()
        threshold_t[node],steps[node]=maximal_perturbation3(L,node,tmin,tmax)
        computation_time[node] = time.time() - t0
    return threshold_t, computation_time,steps
```

```python
nmin = 10
nmax = 100
step = 20
N_iter = 10
all_p = np.arange(0.05,0.3,0.05)
all_results = np.zeros((np.int((nmax-nmin)/step)+1,len(all_p),2))

t0 = time.time()
for i,n in enumerate(range(nmin,nmax,step)):
    print(n)
    for j,p in enumerate(all_p):
        m = np.int(p*n*(n-1)/2)
        all_median_steps = []
        all_median_centrality = []
        print(p)
        for k in range(N_iter):
            G = nx.gnm_random_graph(n, m)
            '''centrality = nx.betweenness_centrality(G)
            b_cent = [v for v in centrality.values()]
            centrality = nx.closeness_centrality(G)
            c_cent = [v for v in centrality.values()]
            centrality = nx.eigenvector_centrality(G)
            e_cent = [v for v in centrality.values()]'''
            L = nx.laplacian_matrix(G)
            L = L.toarray()
            t_centrality,_,steps = threshold_centrality_comp_time(L,1e-10,1e10)
            all_median_steps.append(np.median(steps))
            all_median_centrality.append(np.median(1./t_centrality))
        all_results[i,j,0] = np.mean(all_median_steps)
        all_results[i,j,1] = np.mean(all_median_centrality)


print("Total time = "+str(time.time()-t0))
```
